```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
file_path ="/content/Civil_Engineering_Regression_Dataset.csv"
df = pd.read_csv(file_path)

# Drop Project_ID as it's not a feature
df.drop(columns=["Project_ID"], inplace=True)

# Display the first few rows
print("First few rows of the dataset:")
print(df.head())

# Identify independent and dependent variables
independent_vars = df.columns[df.columns != "Construction_Cost"]
dependent_var = "Construction_Cost"

print("\nIndependent Variables:", independent_vars.tolist())
print("Dependent Variable:", dependent_var)

# Check for missing values
missing_values = df.isnull().sum()
print("\nMissing values in each column:\n", missing_values)

# Generate summary statistics
print("\nSummary statistics:")
print(df.describe())

# Create correlation heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()

# Interpretation of correlation
print("\nCorrelation with Construction Cost:")
print(df.corr()["Construction_Cost"].sort_values(ascending=False))
```
